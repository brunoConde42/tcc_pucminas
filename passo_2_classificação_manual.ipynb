# Classifica√ß√£o manual dos Tweets
Nesta etapa os tweets ser√£o classificados quanto √† polaridade.
Dos 651 tweets 150 ser√£o aleatoriamente selecionados para classifia√ß√£o manual.
Na classifica√ß√°o ser√£o utilizados os seguintes valores: POSITIVO, NEGATIVO, NEUTRO.

```python
import pandas as pd
```

```python
#lendo os tweeds gravados na etapa anterior e carregando em um data frame 
df = pd.read_pickle('dados_twitter_ans.pkl')
df.shape
```

```python
#carregando em excel para classifica√ß√£o
df.to_excel("tweets_sample.xlsx") 
```

```python
#carregando o excel preenchido
df_sample_classificado = pd.read_excel('tweets_sample_preenchidos.xlsx', index_col=0)
df_sample_classificado
```

```python
#criando novas colunas
df['POSITIVO'] = 0
df['NEGATIVO'] = 0
df['NEUTRO'] = 0
df['SPAM'] = 0

# atribuindo os valores
for index, row in df_sample_classificado.iterrows():
    df.at[index,'POSITIVO'] = row['POSITIVO']
    df.at[index,'NEGATIVO'] = row['NEGATIVO']
    df.at[index,'NEUTRO'] = row['NEUTRO']
    df.at[index,'SPAM'] = row['SPAM']

#retirando os SPAMs
df = df[df.SPAM != 1]
df.shape
```

```python
print(len(df[df.POSITIVO == 1]))
print(len(df[df.NEGATIVO == 1]))
print(len(df[df.NEUTRO == 1]))
```

```python
#gravando arquivos
df.to_pickle('dados_twitter_ans_classificados.pkl')
```
